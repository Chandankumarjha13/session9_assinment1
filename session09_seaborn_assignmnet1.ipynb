```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris 
iris = load_iris() 
data = pd. DataFrame(iris.data, columns=iris.feature_names) 
label = pd.DataFrame(list(map(lambda x : iris.target_names[x], iris.target)), columns=['Species']) 
iris = pd.concat([data, label], axis=1) 
print(iris.head()) 
```

1) Use the distplot() to see the distribution of the SepalLengthCm, SepalWidthcm, P etalLengthCm, PetalWidthCm features. Plot them as subplots in a single image.

```python
sns.distplot(iris['sepal length (cm)'], color='skyblue', label='Sepal Length')
sns.distplot(iris['sepal width (cm)'], color='red', label='Sepal Width')
sns.distplot(iris['petal length (cm)'], color='gold', label='petal length')
sns.distplot(iris['petal width (cm)'], color='green', label='petal width')
# sns.legend()
# sns.show()
```

```python
sns.boxplot(data=iris.iloc[:,0:4])
```

```python
sns.countplot('Species', data = iris)
```
